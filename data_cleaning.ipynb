```python
import pandas as pd
```

```python
df = pd.read_csv('letterboxd-reviews.csv', encoding='latin1')
```

```python
df.head()
```

```python
import numpy as np
# Convert 'Rating' to a numerical value
# Replace the star symbols (â??) and half-star symbols (â½) with Unicode characters
# count the characters to get a numerical rating
df['Rating'] = df['Rating'].astype(str).str.replace('â??', '★', regex=False).str.replace('â½', '½', regex=False)
df['Rating'] = df['Rating'].apply(lambda x: x.count('★') + x.count('½') * 0.5 if isinstance(x, str) else np.nan)
print("--- First 5 rows of cleaned 'Rating' column ---")
print(df['Rating'].head())
# And a new statistical summary to see the change
print("\n--- Statistical summary including 'Rating' ---")
print(df.describe())
```

```python
# helper function to clean the numerical columns
def clean_numeric_column(series):
    series = series.astype(str).str.replace(',', '', regex=False).str.strip()
    # Handle 'K' for thousands
    series = series.str.replace(r'(\d+\.?\d*)K', r'\1000', regex=True)
    return pd.to_numeric(series, errors='coerce')
# Apply the cleaning function to the 'Like count' and 'Comment count' columns
df['Like count'] = clean_numeric_column(df['Like count'])
df['Comment count'] = clean_numeric_column(df['Comment count'])
# check the data types to confirm the cleaning was successful
print("--- Final DataFrame Info ---")
print(df.info())
# statistical summary for the cleaned data
print("\n--- Final Statistical Summary ---")
print(df.describe())
```

```python
df.head()
```

```python
import re
# Clean the 'Movie name' column
# We will remove the release year from the movie title for consistency.
df['Movie name'] = df['Movie name'].str.replace(r'\s*\(\d{4}\)\s*$', '', regex=True)
# Clean the 'Review' text
# This removes unwanted special characters from the reviews.
df['Review'] = df['Review'].str.replace(r'[^A-Za-z0-9\s.?!,]', '', regex=True).str.strip()
# Display the final, cleaned DataFrame head to see the results
print("--- Final Cleaned DataFrame Head ---")
print(df.head())
```

```python
from sentence_transformers import SentenceTransformer
# Load a pre-trained embedding model
model = SentenceTransformer('all-MiniLM-L6-v2')
# Explicitly convert the 'Review' column to string to handle any non-string values
df['Review'] = df['Review'].astype(str)
# Create a list of the text to embed
reviews_list = df['Review'].tolist()
# Generate the embeddings
# This might take a few moments depending on the number of reviews
embeddings = model.encode(reviews_list)
# Now, let's add the embeddings back into the DataFrame
df['embeddings'] = list(embeddings)
# Print the shape of the embeddings to confirm they were created
print("Shape of the embeddings array:", embeddings.shape)
# Display the first few rows of the DataFrame with the new embeddings column
print("\n--- DataFrame with Embeddings ---")
print(df.head())





```

```python
import chromadb
import uuid
# 1. Initialize ChromaDB client
# We will use a persistent client, so the data is saved to a folder
client = chromadb.PersistentClient(path="./chroma_db")
# 2. Create or get a collection
# A collection is like a table in a traditional database
collection_name = "movie_reviews"
collection = client.get_or_create_collection(name=collection_name)
# 3. Prepare the data for insertion
# We need three lists: unique ids, the embeddings, and the original text
ids = [str(uuid.uuid4()) for _ in range(len(df))]
documents = df['Review'].tolist()
embeddings_list = df['embeddings'].tolist()
# 4. Add the data to the collection
collection.add(
    ids=ids,
    embeddings=embeddings_list,
    documents=documents
)
print(f"Successfully added {len(df)} reviews to the '{collection_name}' collection.")
print(f"Total documents in collection: {collection.count()}")
# Optional: Perform a quick test query to make sure it works
# This requires the embedding model to be loaded again for the query text
from sentence_transformers import SentenceTransformer
model = SentenceTransformer('all-MiniLM-L6-v2')
test_query = "What is a good comedy movie?"
query_embedding = model.encode([test_query]).tolist()[0]
test_results = collection.query(
    query_embeddings=[query_embedding],
    n_results=3
)
print("\n--- Test Query Results ---")
print(test_results)






```
